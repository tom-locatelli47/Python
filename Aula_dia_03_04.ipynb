```python
class Retangulo:
  def __init__(self,comprimento =10,largura =20):
    self.comprimento = comprimento
    self.largura = largura

  def mudar(self):
    novo_comprimento = int(input(f"Informe o novo comprimento(em cm): "))
    nova_largura = int(input(f"Informe a nova largura(em cm): "))
    self.comprimento = novo_comprimento
    self.largura = nova_largura

  def valor(self):
    return (self.comprimento, self.largura)

  def area(self):
    area = self.comprimento * self.largura
    return area

  def perimetro(self):
    perimetro = (self.comprimento*2) + (self.largura*2)
    return perimetro

retangulo = Retangulo()
```

```python
print(f"Perimetro: {retangulo.perimetro()}cm")
```

```python
print(f"Area: {retangulo.area()}cm")
```

```python
retangulo.mudar()
```

```python
class Pessoa:
  def __init__(self,nome,idade,peso,altura):
    self.nome = nome
    self.idade = idade
    self.peso = peso
    self.altura = altura

  def getnome(self):
    return f"Nome: {self.nome}"

  def getidade(self):
    return f"idade: {self.idade}"

  def getpeso(self):
    return f"Peso: {self.peso}"

  def getaltura(self):
    return f"Altura: {self.altura}"

  def setidade(self,idade):  # poderia colocar o envelhecer aqui
    self.idade

  def setpeso(self,peso): # poderia colocar o engordar aqui
    self.peso

  def setaltura(self,altura): # poderia colocar o crescer aqui
    self.altura

  def main(self):
    return f"Nome: {self.nome}\nIdade: {self.idade}\nPeso: {self.peso}kg\nAltura: {self.altura}cm"

  def envelhecer(self):
    self.idade = self.idade + 1
    if self.idade <= 21:
      self.altura = self.altura + 0.5

  def engordar(self):
    self.peso = self.peso + 1

  def emagrecer(self):
    if self.peso >= 1:
      self.peso = self.peso - 1

    else:
      return self.peso

  def crescer(self):
    self.altura = self.altura + 1



pessoa1=Pessoa("john",17,6,178)

print(pessoa1.main())
```

```python
pessoa1.envelhecer()
pessoa1.engordar()
pessoa1.crescer()
print(pessoa1.main())
```

```python
class Conta:
  def __init__(self,numero,nome,saldo=0):
    self.numero = numero
    self.nome = nome
    self.saldo = saldo

  def getnome(self):
    return f"Nome do correntista: {self.nome}"

  def getsaldo(self):
    return f"Saldo atual: R${self.saldo}"

  def getnumero(self):
    return f"Numero da conta: {self.numero}" # poderia colocar isso no saque e no deposito

  def deposito(self):
    self.saldo = self.saldo + 10

  def saque(self):
    self.saldo = self.saldo - 10

  def alterarnome(self):
    self.nome = str(input("Novo nome: "))

  def main(self):
    return f"Conta: {self.numero}\nNome: {self.nome}\nSaldo: {self.saldo}"

conta1 = Conta(7,"Jonh")

print(conta1.main())
```
