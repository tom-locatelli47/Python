```python
from math import pi

class Forma():
  def area(self):
    pass

class Circulo(Forma):
  def __init__(self,raio):
    self.raio = raio

  def area(self):
    return f"area do circulo: {pi * self.raio**2}"

class Quadrado(Forma):
  def __init__(self,lado):
    self.lado =lado

  def area(self):
    return f"area do quadrado: {self.lado**2}"

circulo1=Circulo(2)
quadrado1=Quadrado(2)

lista = [circulo1,quadrado1]

for formas in lista:
  print(formas.area())
```

```python
class Empregado():
  def pagar_salario(self):
    pass

class EmpregadoHora(Empregado):
  def __init__(self,horas):
    self.horas=horas

  def pagar_salario(self):
    return f"Salario do colaborador que recebe por hora: {12*self.horas}\nEle trabalhou por {self.horas} hora(s)"

class EmpregadoMes(Empregado):
  def __init__(self,meses):
    self.meses = meses

  def pagar_salario(self):
     return f"Salario do colaborador que recebe por mes: {2400*self.meses}\nEle trabalhou por {self.meses} mes/meses"

empregadohora = EmpregadoHora(18)
empregadomes = EmpregadoMes(1)

lista = [empregadohora,empregadomes]

for salarios in lista:
  print(f"{salarios.pagar_salario()}\n")
```

```python
class ContaBancaria():
  def deposito(self):
    pass

  def retirada(self):
    pass

class ContaPoupanca(ContaBancaria):
  def __init__(self,saldo):
    self.saldo = saldo

  def calc_juros(self):
    self.saldo += self.saldo * (10/100)

  def deposito(self,deposito):
    self.saldo = self.saldo + deposito

  def retirada(self,retirada):
    if self.saldo >= retirada:
        self.saldo = self.saldo - retirada

    else:
      return "Você não tem saldo suficiente na conta poupança"

class ContaCorrente(ContaBancaria):
  def __init__(self,saldo):
    self.saldo = saldo

  def calc_juros(self):
    self.saldo += self.saldo * (1/100)

  def deposito(self,deposito):
    self.saldo = self.saldo + deposito

  def retirada(self,retirada):
    if self.saldo >= retirada:
        self.saldo = self.saldo - retirada

    else:
      return "Você não tem saldo suficiente na conta corrente"


poupanca1=ContaPoupanca(1000)
corrente1=ContaCorrente(1000)

lista= [poupanca1,corrente1]
```

```python
for juros in lista:
  juros.calc_juros()
  print(f"{juros.saldo:.2f}")
```
