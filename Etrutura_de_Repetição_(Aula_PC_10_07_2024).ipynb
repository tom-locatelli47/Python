1. Faça um algoritmo que:
• leia 5 números inteiros;
• escreva os números que são negativos;
• escreva a média dos números positivos.

```python
c = 0   #contador
sp = 0  #soma positivo
cp = 0  #contador dos positivos

while c <5:
  num = int(input("Digite um numero: "))
  if num <0:
    print(f"O numero {num} é megativo")
  else:
    sp = sp + num
    cp = cp + 1
  c = c +1

  media = sp / cp
print(f" A Media dos numeros possitivos é {media}")
```

4. Faça um algoritmo que calcule e escreva a soma dos números pares e a soma dos
números ímpares entre 1 e 100.

```python
contador = 0
somapar = 0
somaimpar = 0

while contador <=100:

  if contador%2 == 0:
    somapar = somapar + contador
  else:
    somaimpar = somaimpar + contador

  contador = contador + 1

print(f"a soma dos numeros pares de 1 a 100 é: {somapar}")
print(f"a soma dos numeros impares de 1 a 100 é: {somaimpar}")
```

5. Faça um algoritmo que leia a altura de 20 pessoas e calcule a média aritmética das alturas.

6. Faça um algoritmo que leia n valores inteiros e escreva quantos desses valores são
negativos.

```python
n = int(input("Informe a quantidade de numeros q deseja verificar: "))
contador = 0
negativos = 0

while contador < n:
  num = int(input("Digite o numero q deseja verificar: "))
  if num >= 0:
    print("Esse numero é possitivo")
  else:
    print("Esse numero é negativo")
    negativos = negativos + 1
  contador = contador + 1
print(f"Voce digitou {negativos} numeros negtivos")
```
