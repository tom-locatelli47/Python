```python
matriz=[]
matriz2=[]
linhas =20
coluna=20

print()
for i in range(linhas):
  linha=[]   # oque será a matriz original
  linha2=[]  # só foi usada para poder ser dividida pela matriz original
  linha3=[]  # oque sera a matriz depois da divisão

  for j in range(coluna):
    num= random.randint(1,9)
    linha.append(num)
    linha2.append(num)

  matriz.append(linha) # a matriz original

  for k in range(len(linha2)):
   linha3.append(linha2[k] // linha[i]) # divisão inteira

  matriz2.append(linha3) #matriz depois da divisão

print("----------------MATRIZ----------------")
print()

for i in matriz:
  print(i)

print(f"\n---------Matriz depois da divisão---------")
print()

for i in matriz2:
  print(i)
```

```python
matriz=[]
linhas=4
coluna=5
soma=[]
p=1

for i in range(linhas):
  linha=[]
  linha2=[]
  for j in range(coluna):
    num=random.randint(0,9)
    linha.append(num)
    linha2.append(num)

  matriz.append(linha)
  soma.append(sum(linha2))

print("-------Matriz-------\n")
for i in matriz:
  print(i)

print()
for i in soma:
  print(f"A soma da {p}º linha é: {i}")
  p+=1

print(f"\nA a soma das {p-1} linhas é: {sum(soma)}")
```
