```python
def conversão(fahrenheit):
  celsius=(fahrenheit-32) * 5/9
  return celsius

fahrenheit = int(input("Digite a temperatura em fahrenheit: "))
temperatura_celsius = conversão(fahrenheit)
print(f"\nA temperatura {fahrenheit}°F equivale a {temperatura_celsius}°C")
```

```python
def calculo_hipotenusa(CatetuA,CatetuB):
  hipotenusa = math.sqrt(catetuA**2 + catetuB**2)
  return hipotenusa

catetuA = int(input("Digite o valor do catetuA: "))
catetuB = int(input("Digite o valor do CatetuB: "))
valor = calculo_hipotenusa(catetuA,catetuB)

print(f"\nA ipotenusa é {valor:.2f}")
```

```python
def pesoideal(altura,sexo):

 if sexo == 1:
    peso = (62.1 * altura) - 44.7

 if sexo == 2:
    peso = (72.7 * altura) - 58

 return peso

sexo=int(input("informe o sexo (1-feminino ou 2-masculino): "))
altura= float(input("Informe a altura da pessoa(em metros): "))
resultado = pesoideal(altura,sexo)
print(f"\nO pesso idela da pessoa é {resultado:.2f}")
```

```python
def calculo_poligono(lados,medida):
  perimetro = lados * medida
  area = medida * medida

  if lados == 3:
    msg = f"Triangulo, o perimetro é {perimetro}"

  if lados == 4:
    msg = f"Quadrado, a area é {area}"

  if lados == 5:
    msg = "PENTAGONO"
  return msg

lados=int(input("Quantos lados o poligono tem: "))
medida=int(input("Qual a medida do lado( em cm) do plogino: "))
resultado = calculo_poligono(lados,medida)
print("\n",resultado,"cm")
```
